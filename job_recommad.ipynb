```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = pd.read_csv('job_roles_skills_salary.csv')
```

```python
data
```

## Data Preprocessing

```python
data = data[['Job Role','Required Skills','Expected Salary (INR)']]
```

```python
data
```

```python
data['Required Skills'] = data['Required Skills'].apply(lambda x:[i.replace(" ",'') for i in x.split(',')])
```

```python
data['Required Skills'] = data['Required Skills'].apply(lambda x: " ".join(x))
```

```python
data['Required Skills'] = data['Required Skills'].apply(lambda x:x.lower())
```

```python
data['Required Skills']
```

## Text-To-Vector

```python
# model = CountVectorizer()
# vector = model.fit_transform(data['Required Skills']).toarray()
```

```python
model = TfidfVectorizer()
vector = model.fit_transform(data['Required Skills']).toarray()
```

```python
user_skill = ['html css javascript react node.js git']
```

```python
user_vector = model.transform(user_skill)
```

## Similarity

```python
simil = cosine_similarity(user_vector,vector).flatten()
```

```python
vector
```

```python
data['similarity'] = simil
recommended_jobs = data.sort_values(by='similarity', ascending=False)

print(recommended_jobs[['Job Role', 'similarity']][0:5])
```
